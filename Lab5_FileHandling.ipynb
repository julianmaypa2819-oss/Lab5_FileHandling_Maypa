## Procedure 4: Safe File Operations with Backup

```python
from pathlib import Path
from datetime import datetime
import shutil

# Step 2: Define working directory
backup_dir = Path.home() / "Documents" / "Maypa_Activity_5"
backup_dir.mkdir(exist_ok=True)

# Try It: Create a function that lists only backups for a specific filename
def list_specific_backups(filename):
    print(f"\n--- Backups for {filename} ---")
    # Search for files that start with the filename and contain 'backup'
    search_pattern = f"{Path(filename).stem}_backup*"
    for backup in backup_dir.glob(search_pattern):
        print("-", backup.name)

# Step 3: Define function to write with backup
def write_with_backup(filename, content, mode="w"):
    file_path = backup_dir / filename
    
    # Create backup if file exists
    if file_path.exists():
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        # Try It: Add surname (Maypa) to the backup filename format
        backup_name = f"{file_path.stem}_Maypa_backup_{timestamp}{file_path.suffix}"
        backup_path = file_path.with_name(backup_name)
        shutil.copy2(file_path, backup_path)
        print(f"Backup saved: {backup_path.name}")
    
    # Step 4: Write the content
    # Try It: User can now choose 'w' (overwrite) or 'a' (append) via the 'mode' parameter
    with open(file_path, mode, encoding="utf-8") as file:
        file.write(content)
        print(f"File saved: {file_path.name}")

# Step 5: Run demonstration
print("=== File Operations Demo ===")

# 1. Initial creation
print("\nn1. Creating new file:")
write_with_backup("demo.txt", "Initial content")

# 2. Update (This will trigger the backup)
print("\nn2. Updating file (with backup):")
write_with_backup("demo.txt", "\nUpdated content", mode="a")

# 3. Read back
print("\nn3. Reading file:")
with open(backup_dir / "demo.txt", "r") as f:
    print(f.read())

# 4. Listing backups
list_specific_backups("demo.txt")
```

## Procedure 5: Menu-Driven File Manager

```python
from pathlib import Path
from datetime import datetime
import shutil

# Step 2: Custom working directory as per "Try It"
# Replace 'Maypa' with your actual surname if different
backup_dir = Path.home() / "Documents" / "Maypa_Activity_5"
backup_dir.mkdir(exist_ok=True)

# Step 3: Define the menu-driven file manager
def file_manager():
    file_name = input("Enter filename (e.g., notes.txt): ")
    file_path = backup_dir / file_name

    while True:
        print("\n--- MENU ---")
        print("1. Write to file")
        print("2. Append to file")
        print("3. Read file")
        print("4. Backup file")
        print("5. List all backup files") # Added per "Try It"
        print("6. Exit") # Adjusted number for new option

        choice = input("Choose an option (1-6): ")

        if choice == "1":
            content = input("Enter content to write:\n")
            with open(file_path, "w", encoding="utf-8") as f:
                f.write(content)
            print("File written successfully.")

        elif choice == "2":
            more = input("Enter content to append:\n")
            with open(file_path, "a", encoding="utf-8") as f:
                f.write("\n" + more)
            print("Content appended.")

        elif choice == "3":
            if file_path.exists():
                with open(file_path, "r", encoding="utf-8") as f:
                    print("\nFile Content:\n", f.read())
            else:
                print("File not found.")

        elif choice == "4":
            if file_path.exists():
                timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
                # Try It: Modify backup name to include surname
                backup_file = file_path.with_name(f"{file_path.stem}_Maypa_backup_{timestamp}{file_path.suffix}")
                shutil.copy2(file_path, backup_file)
                print(f"Backup created: {backup_file.name}")
            else:
                print("Cannot backup. File does not exist.")

        elif choice == "5":
            # Try It: Add a menu option to list all backup files
            print(f"\n--- All Backups in {backup_dir.name} ---")
            for backup in backup_dir.glob("*backup*"):
                print("-", backup.name)

        elif choice == "6":
            print("Exiting the file manager.")
            break
        else:
            print("Invalid choice. Please try again.")

# Step 4: Run the menu-driven tool
if __name__ == "__main__":
    file_manager()
```

```python
# Row 20: Merge Two Files
from pathlib import Path

# Personal details
student_id = "2025-0357"

# Define file paths (using your established folder "Maypa_Activity_5")
documents_path = Path.home() / "Documents" / "Maypa_Activity_5"
f1 = documents_path / f"intro_{student_id}.txt"
f2 = documents_path / f"lines_{student_id}.txt"
merged = documents_path / f"merged_{student_id}.txt"

# Ensure both source files exist
if not f1.exists():
    f1.write_text(f"Welcome {student_id} to File Handling in Python!")
    print(f"Sample intro file created for Student ID: {student_id}")

if not f2.exists():
    f2.write_text("Line 1\nLine 2\nLine 3")
    print(f"Sample lines file created for Student ID: {student_id}")

# Merge contents of both files into a new file
with merged.open("w") as mf:
    mf.write(f1.read_text())
    mf.write("\n") # Add a newline between the two files
    mf.write(f2.read_text())

print(f"Files merged successfully for Student ID: {student_id}")
print("--- Merged Content ---")
print(merged.read_text())
```
